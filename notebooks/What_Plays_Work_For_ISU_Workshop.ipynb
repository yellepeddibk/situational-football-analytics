```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re # Added for text parsing in scoring plays

# --- Global Definitions and Helper Functions ---
# Define play categories here for use in this cell
rushing_plays = ['Rush', 'Rushing Touchdown']
passing_plays = ['Pass Reception', 'Passing Touchdown', 'Pass Incompletion', 'Sack']

# Define the categorization function
def categorize_play(play_type):
    if play_type in rushing_plays:
        return 'Rushing'
    elif play_type in passing_plays:
        return 'Passing'
    else:
        return 'Other'

# Function to classify play type based on the user's specific criteria for scatter_box_run_vs_pass
def classify_user_play(playType):
    if playType in rushing_plays:
        return 'Run'
    elif playType in passing_plays:
        return 'Pass'
    else:
        return 'Other' # For plays like Kickoffs, Punts, Penalties, etc.

# --- Plotting Functions ---

def yards_gained_by_distance(df, opponent_name):
    # Filter for Redzone plays (yardsToGoal <= 20)
    df_redzone = df[df['yardsToGoal'] <= 20].copy()
    print(f"{opponent_name} Redzone Plays shape: {df_redzone.shape}")

    # Filter for only Rushing and Passing plays in the Redzone to define df_yards
    df_yards = df_redzone[df_redzone['playType'].isin(rushing_plays + passing_plays)].copy()

    # Apply the play category
    df_yards['playCategory'] = df_yards['playType'].apply(categorize_play)

    # Calculate the average yardsGained for each category
    average_yards = df_yards.groupby('playCategory')['yardsGained'].mean().reset_index()
    average_yards.rename(columns={'yardsGained': 'Average Yards Gained'}, inplace=True)

    print(f"\nAverage Yards Gained in Redzone (vs {opponent_name}):")
    print(average_yards.to_markdown(index=False, numalign="left", stralign="left"))
    print("-" * 50)

    # Create the Scatterplot
    plt.figure(figsize=(8, 6))

    if not average_yards.empty:
        # Map playCategory to a numerical x-axis value (0 for Rushing, 1 for Passing)
        x_values = [0 if cat == 'Rushing' else 1 for cat in average_yards['playCategory']]
        plt.scatter(x_values, average_yards['Average Yards Gained'], s=200, color=['firebrick', 'cornflowerblue'])

        # Set x-axis labels to the play categories
        plt.xticks([0, 1], average_yards['playCategory'], fontsize=12)

        # Add annotations to the points
        for i, row in average_yards.iterrows():
            plt.annotate(
                f"{row['Average Yards Gained']:.2f}",
                (x_values[i], row['Average Yards Gained']),
                textcoords="offset points", # how to position the text
                xytext=(0,10), # distance from text to points (x,y)
                ha='center', # horizontal alignment
                fontsize=10
            )

        plt.title(f'Average Yards Gained in the Redzone for Iowa State Offense vs {opponent_name}', fontsize=14)
        plt.xlabel('Play Category', fontsize=12)
        plt.ylabel('Average Yards Gained', fontsize=12)
        plt.ylim(bottom=0) # Start the y-axis at 0
        plt.grid(axis='y', linestyle='--')
        plt.savefig(f'redzone_avg_yards_scatterplot_{opponent_name.lower().replace(" ", "_")}.png')
        plt.show()
    else:
        print(f"No data available to plot Average Yards Gained in Redzone for {opponent_name}.")
        plt.close() # Close the empty figure to prevent it from being displayed

def scatter_box_run_vs_pass(df, opponent_name):
    # Add the new column to the offense DataFrame
    df['UserDefinedPlayType'] = df['playType'].apply(classify_user_play)

    # Create df_rushing and df_passing
    df_rushing = df[df['UserDefinedPlayType'] == 'Run'].copy()
    df_passing = df[df['UserDefinedPlayType'] == 'Pass'].copy()

    print(f"\nSuccessfully created DataFrames for {opponent_name} offensive, passing, and rushing plays.")

    # 1. Filter for 1st, 2nd, and 3rd down
    down_filter = [1, 2, 3]
    df_rushing_filtered = df_rushing[df_rushing['down'].isin(down_filter)]
    df_passing_filtered = df_passing[df_passing['down'].isin(down_filter)]

    # 2. Define the down order and labels for the X-axis
    down_order = [1, 2, 3]
    down_labels = [f'{d}st' if d==1 else (f'{d}nd' if d==2 else f'{d}rd') for d in down_order]

    # 3. Create a figure with two subplots side-by-side, sharing the Y-axis
    fig, axes = plt.subplots(1, 2, figsize=(14, 7), sharey=True)
    fig.suptitle(f'Iowa State Yards Gained vs {opponent_name}: Run vs Pass by Down (Box & Scatter)', fontsize=16)

    # --- Subplot 1: Running Plays (Box and Scatter) ---
    ax1 = axes[0]
    sns.boxplot(
        data=df_rushing_filtered,
        x='down',
        y='yardsGained',
        ax=ax1,
        order=down_order,
        color='#ADD8E6', # Light blue for box
        linewidth=1.5,
        showfliers=False # Hide standard boxplot outliers since we'll plot them with stripplot
    )
    # Overlay the individual run plays as circles ('o')
    sns.stripplot(
        data=df_rushing_filtered,
        x='down',
        y='yardsGained',
        ax=ax1,
        order=down_order,
        color='blue', # Blue for markers
        marker='o',
        s=8, # size of the markers
        alpha=0.6,
        jitter=0.2 # Spread points horizontally for visibility
    )

    # Set labels for the Rushing Plot
    ax1.set_title('Rushing Plays (Run: Circles)', fontsize=14)
    ax1.set_xlabel('Down', fontsize=12)
    ax1.set_ylabel('Yards Gained', fontsize=12)
    ax1.set_xticks(range(len(down_order)))
    ax1.set_xticklabels(down_labels)
    ax1.axhline(0, color='gray', linestyle='--', linewidth=0.8)
    ax1.grid(axis='y', linestyle=':', alpha=0.6)


    # --- Subplot 2: Passing Plays (Box and Scatter) ---
    ax2 = axes[1]
    sns.boxplot(
        data=df_passing_filtered,
        x='down',
        y='yardsGained',
        ax=ax2,
        order=down_order,
        color='#FFA07A', # Light salmon for box
        linewidth=1.5,
        showfliers=False
    )
    # Overlay the individual pass plays as X's ('X')
    sns.stripplot(
        data=df_passing_filtered,
        x='down',
        y='yardsGained',
        ax=ax2,
        order=down_order,
        color='red', # Red for markers
        marker='X',
        s=10, # size of the markers
        alpha=0.7,
        jitter=0.2
    )

    # Set labels for the Passing Plot
    ax2.set_title('Passing Plays (Pass: X\'s)', fontsize=14)
    ax2.set_xlabel('Down', fontsize=12)
    ax2.set_ylabel('') # Clear Y-axis label since it's shared
    ax2.set_xticks(range(len(down_order)))
    ax2.set_xticklabels(down_labels)
    ax2.axhline(0, color='gray', linestyle='--', linewidth=0.8)
    ax2.grid(axis='y', linestyle=':', alpha=0.6)

    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make room for suptitle
    plt.savefig(f'scatter_box_run_vs_pass_{opponent_name.lower().replace(" ", "_")}.png')
    plt.show()

def plot_all_redzone_plays_yards_gained(df_opponent, opponent_name):
    # Filter for Redzone plays (yardsToGoal <= 20)
    df_redzone = df_opponent[df_opponent['yardsToGoal'] <= 20].copy()

    # Sort by 'id' to ensure chronological order for the plot
    df_redzone.sort_values(by='id', inplace=True)

    # Add a chronological play index
    df_redzone['Play Index'] = range(len(df_redzone))

    # Add play category
    df_redzone['playCategory'] = df_redzone['playType'].apply(categorize_play)

    # Create the Scatterplot
    plt.figure(figsize=(12, 6))

    # Define colors for play categories
    category_colors = {'Rushing': 'firebrick', 'Passing': 'cornflowerblue', 'Other': 'gray'}
    colors = df_redzone['playCategory'].map(category_colors)

    # Scatter plot
    plt.scatter(
        df_redzone['Play Index'],
        df_redzone['yardsGained'],
        c=colors, # Color based on the mapped colors
        s=100,
        alpha=0.8
    )

    # Add legend manually
    legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=category,
                               markerfacecolor=color, markersize=10)
                      for category, color in category_colors.items()]
    plt.legend(handles=legend_handles, title='Play Category')

    # Add a horizontal line at 0 yards gained
    plt.axhline(0, color='black', linestyle='--', linewidth=1)

    # Set labels and title
    plt.title(f'All Redzone Plays: Yards Gained for Iowa State Offense vs {opponent_name}', fontsize=14)
    plt.xlabel('Redzone Play Sequence (Chronological Order)', fontsize=12)
    plt.ylabel('Yards Gained', fontsize=12)

    # Set x-axis ticks to be manageable
    max_index = df_redzone['Play Index'].max()
    if max_index > 0:
        plt.xticks(range(0, max_index + 1, max_index // 5 or 1)) # show about 5-6 ticks
    else:
        plt.xticks([]) # No ticks if no plays

    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()
    plt.savefig(f'all_redzone_plays_scatterplot_{opponent_name.lower().replace(" ", "_")}.png')
    plt.show()

def yards_per_play_per_qtr(df, opponent_name):
    # Calculate the average yards per play for each quarter
    if not df.empty:
        avg_yards_per_play = df.groupby('period')['yardsGained'].mean().reset_index()
        avg_yards_per_play.columns = ['period', 'Avg Yards Per Play']

        print(f"\nCalculated Average Yards Per Play by Quarter (vs {opponent_name}):\n{avg_yards_per_play}")

        # Create the bar plot
        plt.figure(figsize=(10, 6))

        sns.barplot(
            x='period',
            y='Avg Yards Per Play',
            data=avg_yards_per_play,
            palette='magma'
        )

        # Add labels and a title
        plt.title(f'Iowa State Offensive Average Yards Per Play vs. {opponent_name} by Quarter')
        plt.xlabel('Quarter')
        plt.ylabel('Average Yards Gained')

        # Add gridlines
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Display the value on top of each bar
        for index, row in avg_yards_per_play.iterrows():
            plt.text(
                index,
                row['Avg Yards Per Play'] + 0.1,
                f"{row['Avg Yards Per Play']:.2f}",
                color='black',
                ha='center'
            )
        plt.savefig(f'avg_yards_per_play_per_qtr_{opponent_name.lower().replace(" ", "_")}.png')
        plt.show()
    else:
        print(f"No data available for Iowa State offense vs {opponent_name} to calculate average yards per play by quarter.")

def plot_yards_gained_by_down_distance(df, opponent_name):
    # 1. Define bins and labels for the distance column (intervals of 3 yards)
    bins = [1, 4, 7, 10, 13, 16]
    labels = ['1-3', '4-6', '7-9', '10-12', '13-15']

    # 2. Create the grouped distance column
    df['distance_group'] = pd.cut(
        df['distance'],
        bins=bins,
        labels=labels,
        right=False,
        include_lowest=True
    )

    # 3. Group by 'down' and 'distance_group', calculate the average yards gained
    df_avg_yards_down_grouped = df[df['down'].isin([1, 2, 3])].groupby(
        ['down', 'distance_group'],
        observed=True,
        as_index=False
    )['yardsGained'].mean()

    # 4. Prepare data for plotting
    df_avg_yards_down_grouped.rename(columns={'yardsGained': 'Average Yards Gained'}, inplace=True)
    df_avg_yards_down_grouped['down_str'] = df_avg_yards_down_grouped['down'].astype(str)

    # Print the summary table
    print("\n" + "="*70)
    print(f"Average Yards Gained by Down and Distance Interval (vs. {opponent_name}):")
    print(df_avg_yards_down_grouped.to_markdown(index=False, numalign="left", stralign="left"))
    print("="*70 + "\n")

    # 5. Create the grouped bar plot using Seaborn/Matplotlib
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(10, 6))

    # Define the custom legend labels with ordinal suffixes
    down_labels = {
        '1': '1st Down',
        '2': '2nd Down',
        '3': '3rd Down'
    }

    barplot = sns.barplot(
        data=df_avg_yards_down_grouped,
        x='distance_group',
        y='Average Yards Gained',
        hue='down_str',
        order=labels,
        errorbar=None
    )

    # Apply custom labels and title
    plt.title(f'Iowa State Offense: Avg. Yards Gained by Down within Distance Intervals (vs. {opponent_name})', fontsize=14)
    plt.xlabel('Yards to First Down Interval', fontsize=12)
    plt.ylabel('Average Yards Gained on Play', fontsize=12)

    # Customize the legend to use the ordinal suffixes
    handles, original_labels = barplot.get_legend_handles_labels()
    custom_labels = [down_labels[lab] for lab in original_labels]
    plt.legend(handles, custom_labels, title='Down', loc='upper left')

    # Save and display the plot
    plt.savefig(f'avg_yards_gained_by_down_distance_{opponent_name.lower().replace(" ", "_")}.png')
    plt.show()

# --- Main Execution Block ---

# 1. Get the path to the CSV file.
file_path = '/content/iowa_state_2025_full_plays.csv'

try:
    # 2. Read the CSV file into a Pandas DataFrame
    df_raw = pd.read_csv(file_path)

    # ------------------------------
    # --- DATA CLEANING: Drop Duplicates ---
    initial_rows = len(df_raw)
    df_raw.drop_duplicates(keep='first', inplace=True)
    rows_dropped = initial_rows - len(df_raw)
    print(f"Data Cleaning: Dropped {rows_dropped} duplicate rows across all columns.")
    # ------------------------------

    print("Successfully loaded the Play-by-Play Data!")
    print(f"Raw data shape: {df_raw.shape}")

    # 3. Filter the entire dataset to only include plays where Iowa State is on offense
    df_is_offense = df_raw[df_raw['offense'] == 'Iowa State'].copy()
    print(f"Filtered to Iowa State Offense only. New shape: {df_is_offense.shape}")
    print("\n" + "="*70 + "\n")

    # 4. Create all opponent DataFrames in a dictionary
    # Get unique opponent names from the 'defense' column, excluding 'Iowa State'
    unique_defenses = df_is_offense['defense'].unique().tolist()
    if 'Iowa State' in unique_defenses:
        unique_defenses.remove('Iowa State')

    opponent_dfs = {}
    for name in unique_defenses:
        opponent_dfs[name] = df_is_offense[df_is_offense['defense'] == name].copy()

    # 5. Loop through the opponent DataFrames and call all plotting functions for each
    for opponent_name, df_opponent in opponent_dfs.items():
        if df_opponent.empty:
            print(f"No data for Iowa State offense vs {opponent_name}. Skipping plots.")
            continue

        print(f"\n--- Processing {opponent_name} ---")
        # Pass a copy to each function to ensure operations within the function are isolated
        yards_gained_by_distance(df_opponent.copy(), opponent_name)
        scatter_box_run_vs_pass(df_opponent.copy(), opponent_name)
        plot_all_redzone_plays_yards_gained(df_opponent.copy(), opponent_name)
        yards_per_play_per_qtr(df_opponent.copy(), opponent_name)
        plot_yards_gained_by_down_distance(df_opponent.copy(), opponent_name)

except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}. Please check the path.")
except Exception as e:
    print(f"An error occurred: {e}")
```
